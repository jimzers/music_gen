```python


#todo: the notes all get appended to one long array.
#but this considers each different song as one big long song. this is wrong!

notes = []

# parse all the files with .mid extension
for file in glob.glob("midi/*.mid"):
    
    # load file into a Music21 stream obj
    midi = converter.parse(file)
    notes_to_parse = None
    
    # get all the notes and chords in the file
    parts = instrument.partitionByInstrument(midi)
    #print(parts)
    
    # get notes to parse 
    if parts:
        # if file has instrument parts
        notes_to_parse = parts.parts[0].recurse()
    else:
        # note are in a flat structure
        notes_to_parse = midi.flat.notes
        
    for element in notes_to_parse:
        
        if isinstance(element, note.Note):
            '''
            print("printing element")
            print(str(element.pitch))
            '''
            notes.append(str(element.pitch))
        elif isinstance(element, chord.Chord):
            '''
            print("alternative print")
            print(element.normalOrder)
            '''
            notes.append('.'.join(str(n) for n in element.normalOrder))
     
    
n_vocab = len(set(notes))
print(n_vocab)
#notes
```

```python
def make_one_hot(width, position):
    
    temp_arr = np.zeros((width,))
    temp_arr[position] = 1
    return temp_arr


make_one_hot(4, 1)
```

```python
# make all the items


# get all pitch names
pitchnameList = set(notes)

one_hot_dim = len(pitchnameList)

# create a dicty to map pitches to integers
note_to_int = dict((note, number) for number, note in enumerate(pitchnameList, 1))
# padding thing should map to null val
note_to_int['PADDING'] = 0


# create one hot encoding for null value
padding_one_hot = np.zeros((one_hot_dim,))

#int to one hot enc
int_to_one_hot = dict((number, make_one_hot(one_hot_dim, number-1)) for number in range(1, one_hot_dim+1))

#reverse the encodings
int_to_note = dict((number, note) for number, note in enumerate(pitchnameList, 1))
int_to_note[0] = 'PADDING'

one_hot_to_int = {}
for integer, one_hot in int_to_one_hot.items():
    one_hot_to_int[one_hot] = integer

one_hot_to_int
```

```python


# create input sequences for the network and the respective outputs

sequence_length = 100

# get all pitch names
pitchnames = sorted(set(item for item in notes))

# create a dicty to map pitches to integers
note_to_int = dict((note, number) for number, note in enumerate(pitchnames))
print(note_to_int)

int_to_one_hot = "poo"




network_input = []
print(type(network_input))
network_output = []

# create input seq3uences and the corresponding outputs
for i in range(0, len(notes) - sequence_length, 1):
    '''
    print("data type of notes")
    print(type(network_input))
    print(network_input)
    print("cycle begins here ^^^")
    '''
    # the variable length input data sequence
    sequence_in = notes[i:i + sequence_length]
    '''
    print("sequence in length")
    print(len(sequence_in))
    print(sequence_in)
    print("POPOPOPOP")
    '''
    # the next input in the given sequence - ground truth data
    sequence_out = notes[i + sequence_length]
    
    
    
    # translate the notes into integer label format, this is X data
    network_in_arr = [note_to_int[char] for char in sequence_in]
    #print(network_in_arr)
    #print("^^^ network in arr")
    network_input.append([note_to_int[char] for char in sequence_in])
    # this is y data
    network_output.append(note_to_int[sequence_out])

# number for training examples
n_patterns = len(network_input)

# reshape the input into a format LSTM layer can use
# (input_length, time steps, time steps of output var)
network_input = np.reshape(network_input, (n_patterns, sequence_length, 1))
# normalize input

print(network_input.shape)

#print(sum(network_input))

network_input = network_input / float(n_vocab)

#print(sum(network_input))

# one hot encode
network_output = to_categorical(network_output)

```

```python
print(network_output[3000])
print(network_input[1])
```

```python
model = Sequential()
model.add(LSTM(
    256,
    input_shape=(network_input.shape[1], network_input.shape[2]),
    return_sequences=True
))
model.add(Dropout(0.3))
model.add(LSTM(512, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(256))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')
```

```python
filepath = "weights.hdf5"    
checkpoint = ModelCheckpoint(
    filepath, monitor='loss', 
    verbose=0,        
    save_best_only=True,        
    mode='min'
)    
callbacks_list = [checkpoint]     
model.fit(network_input, network_output, epochs=1, batch_size=64, callbacks=callbacks_list)

print("finished training")
```

```python
# call up model again

model = Sequential()
model.add(LSTM(
    512,
    input_shape=(network_input.shape[1], network_input.shape[2]),
    return_sequences=True
))
model.add(Dropout(0.3))
model.add(LSTM(512, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')
# Load the weights to each node
model.load_weights('weights.hdf5')
```

```python
offset = 0
output_notes = []
# create note and chord objects based on the values generated by the model
for pattern in prediction_output:
    # pattern is a chord
    if ('.' in pattern) or pattern.isdigit():
        notes_in_chord = pattern.split('.')
        notes = []
        for current_note in notes_in_chord:
            new_note = note.Note(int(current_note))
            new_note.storedInstrument = instrument.Piano()
            notes.append(new_note)
        new_chord = chord.Chord(notes)
        new_chord.offset = offset
        output_notes.append(new_chord)
    # pattern is a note
    else:
        new_note = note.Note(pattern)
        new_note.offset = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
    # increase offset each iteration so that notes do not stack
    offset += 0.5
```
